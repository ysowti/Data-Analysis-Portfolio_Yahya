```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv('data.csv')
```

```python
df['y'].value_counts()
```

```python
count_labels = pd.value_counts(df['y'], sort=True)
count_labels.plot(kind='bar')
plt.title('labels')
plt.xticks(range(2))
plt.xlabel('classes')
plt.ylabel('frequency')
```

```python
for feature in df.columns[:-1]:
    ax = plt.subplot()
    sns.distplot(df[df['y'] == 1][feature], bins=50, label = 'on')
    sns.distplot(df[df['y'] == 0][feature], bins=50, label = 'off')
    ax.set_xlabel('')
    ax.set_title('histogram of features: ' + str(feature))
    plt.legend(loc='best')
    plt.show()
```

```python
fig, ax = plt.subplots(2,1,figsize=(20,12))
sns.boxplot(y = 'x1', x = 'y', hue = 'y', data = df, ax = ax[0])
sns.boxplot(y = 'x2', x = 'y', hue = 'y', data = df, ax = ax[1])
```

```python
num_cols = list(df.columns)[:-1]
plt.figure(figsize = (10,40))
for i, col in enumerate(num_cols):
    ax = plt.subplot(len(num_cols), 1, i + 1)
    sns.scatterplot(x=df[col], y = df['y'], ax=ax)
```
